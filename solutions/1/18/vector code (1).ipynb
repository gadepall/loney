```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x vector contains x1,x2
x=np.array([-3,-8])

# y vector contains y1,y2 
y=np.array([-4,7])
```

```python

m=7
n=5
p=float(m/(m+n))
q=float(n/(m+n))
```

```python
a_arr=q*np.array([-3,-4])
b_arr=p*np.array([-8,7])
```

```python
# Internal coordinate point 

import numpy as np
out_arr= np.add(a_arr, b_arr)
```

```python
print("Internal coordinate points:", out_arr )
```

```python
r=float(m/(m-n))
s=float(n/(m-n))
```

```python
c_arr = r*np.array([-8,7])
d_arr = s*np.array([-3,-4])
```

```python
# External coordinate point

out_arr2= np.subtract(c_arr, d_arr)
```

```python
print("External coordinate points:", out_arr2)
```

```python
import matplotlib.pyplot as plt
x=(-3,-71/12,-8,-41/2)
y=(-4,29/12,7,69/2)
plt.plot(x,y)
plt.plot(x,y,'ro')
plt.plot(-71/12,29/12,'bo',label='Internal divided point S')
plt.plot(-41/2,69/2,'yo',label='External divided point T')
plt.xlabel('x - axis')
plt.ylabel('y-asis')
plt.title('Internally and externally divided coordinate')
for i, j in zip(x, y):
 plt.text(i, j+0.5, '({}, {})'.format(i, j))
plt.grid()
plt.legend()
plt.show()
```
