```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-2,8,100)
y= x-5
plt.plot(x,y,'r', label = 'y=x-5')
plt.title('graph of y =x-5')
plt.legend(loc = 'upper left')
plt.grid()
plt.show()

```
